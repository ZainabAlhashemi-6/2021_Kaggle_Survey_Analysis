```python
preferred_lng = ['Q7_Part_1','Q7_Part_2','Q7_Part_3','Q7_Part_4','Q7_Part_5','Q7_Part_6','Q7_Part_7','Q7_Part_8','Q7_Part_9','Q7_Part_10','Q7_Part_11','Q7_Part_12']
survey_df[preferred_lng]
```

```python
choices = {}

for col in preferred_lng:
    non_null_values = survey_df[col].dropna()
    if not non_null_values.empty:
        full_text = non_null_values.iloc[0]
        answer = full_text.split(' - Selected Choice - ')[-1]
        choices[col] = answer
    else:
        choices[col] = None

print(choices)
```

```python
survey_df.rename(columns=choices, inplace=True)
survey_df
```
