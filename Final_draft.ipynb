## Importing & Understanding the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'kaggle_survey_2021_responses.csv'
survey_df = pd.read_csv(file_path)
survey_df.head(5)
```

```python
survey_df.info()
```

```python
survey_df.columns
```

```python
survey_df[['Q1', 'Q2', 'Q3', 'Q4', 'Q5','Q6', 'Q7_Part_1', 'Q7_Part_2', 'Q7_Part_3','Q7_Part_4','Q7_Part_5','Q7_Part_6','Q7_Part_7','Q7_Part_8','Q7_Part_9','Q7_Part_10','Q7_Part_11','Q7_Part_12']].head(3)
```

```python
survey_df.rename(columns = {'Q1':'age',
                            'Q2':'gender',
                            'Q3':'location',
                            'Q4':'education level',
                            'Q5':'job title',
                            'Q6':'years of experience',
                            'Q7_Part_1':'Python',
                            'Q7_Part_2':'R',
                            'Q7_Part_3':'SQL',
                            'Q7_Part_4':'C',
                            'Q7_Part_5':'C++',
                            'Q7_Part_6': 'Java',
                            'Q7_Part_7': 'Javascript',
                            'Q7_Part_8': 'Julia',
                            'Q7_Part_9': 'Swift',
                            'Q7_Part_10': 'Bash',
                            'Q7_Part_11': 'MATLAB',
                            'Q7_Part_12': 'None'}, inplace=True)
survey_df
```

## Age Analysis of Kaggle Users

### Age Distribution of Kaggle Users

```python
plt.figure(figsize=(10,6))
sns.countplot(survey_df, x='age', order=survey_df['age'].value_counts().index)
plt.title('Age Distribution of Kaggle Users')
plt.xlabel('Age Range')
plt.ylabel('Number of Respondents')
plt.show()
```

### Top 5 Age Groups

```python
age_counts = survey_df['age'].value_counts().sort_values(ascending=False)
top_5_ages = age_counts[:5]

plt.figure(figsize=(8,5))
sns.barplot(x=top_5_ages.values, y=top_5_ages.index)
plt.title('Top 5 Age Groups of Kaggle Users')
plt.xlabel('Number of Respondents')
plt.ylabel('Age Group')
plt.tight_layout()
plt.show()
```

## Educational Level of Kaggle Users

```python
# 5. Education Level
plt.figure(figsize=(10, 6))
edu_counts = survey_df.iloc[1:]['education level'].value_counts()
sns.barplot(x=edu_counts.values, y=edu_counts.index)
plt.title("Formal Education of Kaggle Users")
plt.xlabel("Number of Respondents")
plt.ylabel("Education Level")
plt.tight_layout()
plt.show()
```

## Location Analysis of Kaggle Users

### Top 15 Coutries

```python
top_countries = survey_df['location'].value_counts().head(15)

plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 15 Countries by Number of Participants')
plt.xlabel('Number of Respondents')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Top 5 Countries

```python
top_countries = survey_df['location'].value_counts().head(5)

plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 5 Countries by Number of Participants')
plt.xlabel('Number of Respondents')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

## Job Analysis

### Job Titles of Kaggle Users in 2021

```python
job_counts = survey_df['job title'].value_counts().sort_values(ascending=False)
top_jobs = job_counts

plt.figure(figsize=(10,6))
sns.barplot(x=top_jobs.values, y=top_jobs.index)
plt.title('Job Titles of Kaggle Users (2021)')
plt.xlabel('Number of Respondents')
plt.ylabel('Job Title')
plt.tight_layout()
plt.show()
```

### Top 10 Job titles

```python
job_counts = survey_df['job title'].value_counts().sort_values(ascending=False)
top_jobs = job_counts.head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_jobs.values, y=top_jobs.index)
plt.title('Top 10 Job Titles of Kaggle Users (2021)')
plt.xlabel('Number of Respondents')
plt.ylabel('Job Title')
plt.tight_layout()
plt.show()
```

### Job Titles of Kaggle Users & their Years of Experience

```python
survey_df['years of experience'].unique()
```

```python
job_exp_df = survey_df[['job title','years of experience']].dropna()
top_jobs = survey_df['job title'].value_counts().head().index
filtered = job_exp_df[job_exp_df['job title'].isin(top_jobs)]

plt.figure(figsize=(12,6))
sns.countplot(data=filtered, y='job title', hue='years of experience')
plt.title('Job Titles by Years of Coding Experience (Top 5 Roles)')
plt.xlabel('Number of Respondents')
plt.ylabel('Job Title')
plt.legend(title='Years of Experience')
plt.tight_layout()
plt.show()
```

### Job Titles of Kaggle Users by their Age Groups

```python
top_jobs = survey_df['job title'].value_counts().head().index
filtered = survey_df[survey_df['job title'].isin(top_jobs)]

plt.figure(figsize=(12,6))
sns.countplot(data=filtered, y='job title', hue='age')
plt.title('Job Titles by Age Group')
plt.xlabel('Number of Respondents')
plt.ylabel('Job Title')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Compensation Analysis

### Yearly Compensation of Kaggle Users (USD)

```python
survey_df[['Q25']]
```

```python
salary = survey_df[['Q25']].iloc[1:].dropna()
salary['Q25'].value_counts()
```

```python
salary_count = salary['Q25'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=salary_count.values, y=salary_count.index)
plt.title('Yearly Compensation of Kaggle Users (USD)')
plt.xlabel('Number of Respondents')
plt.ylabel('Salary Range')
plt.tight_layout()
plt.show()
```

### Top 10 Salary Ranges

```python
salary_count = salary['Q25'].head(10).value_counts().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=salary_count.values, y=salary_count.index)
plt.title('Yearly Compensation of Kaggle Users (USD)')
plt.xlabel('Number of Respondents')
plt.ylabel('Salary Range')
plt.tight_layout()
plt.show()
```

### Salary Distribution by Job Titles of Kaggle Users

```python
top_jobs = survey_df['job title'].value_counts().head().index
filtered = survey_df[survey_df['job title'].isin(top_jobs)]

plt.figure(figsize=(12,6))
sns.countplot(data=filtered, y='job title', hue='Q25')
plt.title('Salary Distribution by Job Title')
plt.xlabel('Number of Respondents')
plt.ylabel('Job Title')
plt.legend(title='Yearly Compensation', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
salary_job_df = survey_df[['job title', 'Q25']].dropna()
salary_job_df
```

```python
salary_job_df = salary_job_df[salary_job_df['job title'] != 'Student']
top_jobs = salary_job_df['job title'].value_counts().head(3).index
filtered = salary_job_df[survey_df['job title'].isin(top_jobs)]

plt.figure(figsize=(12,6))
sns.countplot(data=filtered, y='job title', hue='Q25')
plt.title('Salary Distribution by Job Title (Top 3 Roles)')
plt.xlabel('Number of Respondents')
plt.ylabel('Job Title')
plt.legend(title='Salary Range', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## Most Used Programming Languages by Kaggle Users

```python
survey_df = survey_df.iloc[1:].reset_index(drop=True)

languages = survey_df[['Python', 'R', 'SQL', 'C', 'C++', 'Java', 'Javascript', 'Julia', 'Swift', 'Bash', 'MATLAB', 'None']].melt(value_name='Language')['Language'].dropna()

language_counts = languages.value_counts().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=language_counts.values, y=language_counts.index)
plt.title('Most Used Programming Languages')
plt.xlabel('Number of Users')
plt.ylabel('Programming Language')
plt.tight_layout()
plt.show()
```

### Top 5 Programming Languages

```python
survey_df = survey_df.iloc[1:].reset_index(drop=True)

languages = survey_df[['Python', 'R', 'SQL', 'C', 'C++', 'Java', 'Javascript', 'Julia', 'Swift', 'Bash', 'MATLAB', 'None']].melt(value_name='Language')['Language'].dropna()

language_counts = languages.value_counts().head(5).sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=language_counts.values, y=language_counts.index)
plt.title('Most Used Programming Languages')
plt.xlabel('Number of Users')
plt.ylabel('Programming Language')
plt.tight_layout()
plt.show()
```
