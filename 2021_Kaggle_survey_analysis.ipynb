# **`2021 Kaggle Survey Analysis`**

```python
## Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Loading Data set
survey_df = pd.read_csv("kaggle_survey_2021_responses.csv")
survey_df
```

```python
# Rename columns

survey_df.rename(columns={
    'Q1': 'age',
    'Q2': 'gender',
    'Q3': 'location',
    'Q4': 'education level',
    'Q5': 'job title',
    'Q6': 'years of experience',
    'Q11': 'platform',
    'Q20': 'industry',
    'Q25': 'salary'
}, inplace=True)
```

```python
survey_df
```

```python
# Set Seaborn theme
sns.set(style="darkgrid")
```

## `Age Distribution of Kaggle Users`

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=survey_df, x='age', order=survey_df.iloc[1:]['age'].value_counts().index)
plt.title("Age Distribution of Kaggle Users")
plt.xlabel("Age Group")
plt.ylabel("Number of Respondents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Top 5 Age Groups

```python
age_counts = survey_df['age'].value_counts().sort_values(ascending=False)
top_5_ages = age_counts[:5]

plt.figure(figsize=(8,5))
sns.barplot(x=top_5_ages.values, y=top_5_ages.index)
plt.title('Top 5 Age Groups of Kaggle Users')
plt.xlabel('Number of Respondents')
plt.ylabel('Age Group')
plt.tight_layout()
plt.show()
```

## `Education Level of Kaggle Users`

```python
plt.figure(figsize=(10, 6))
edu_counts = survey_df.iloc[1:]['education level'].value_counts()
sns.barplot(x=edu_counts.values, y=edu_counts.index)
plt.title("Formal Education of Kaggle Users")
plt.xlabel("Number of Respondents")
plt.ylabel("Education Level")
plt.tight_layout()
plt.show()
```

## `Country of Residence`

```python
plt.figure(figsize=(14, 8))
top_countries = survey_df['location'].value_counts().head(15)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 15 Countries of Kaggle Respondents")
plt.xlabel("Number of Respondents")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

### Top 5 Countries

```python
plt.figure(figsize=(14, 8))
top_countries = survey_df['location'].value_counts().head(5)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 5 Countries of Kaggle Respondents")
plt.xlabel("Number of Respondents")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

## `Job Titles of Kaggle Users`

```python
plt.figure(figsize=(12, 8))
job_counts = survey_df[1:]['job title'].value_counts()
sns.barplot(x=job_counts.values, y=job_counts.index)
plt.title("Job Titles of Kaggle Users")
plt.xlabel("Number of Respondents")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
```

### `Top 5 Job Titles`

```python
plt.figure(figsize=(12, 8))
job_counts = survey_df[1:]['job title'].value_counts().head()
sns.barplot(x=job_counts.values, y=job_counts.index)
plt.title("Top  Job Titles of Kaggle Users")
plt.xlabel("Number of Respondents")
plt.ylabel("Job Title")
plt.tight_layout()
plt.show()
```

## `Years of Coding Experience`

```python
survey_df['years of experience'].unique()
```

```python
plt.figure(figsize=(10, 6))
exp_order = ['I have never written code', '< 1 years', '1-3 years', '3-5 years',
             '5-10 years', '10-20 years', '20+ years']
sns.countplot(data=survey_df, y='years of experience', order=exp_order)
plt.title("Years of Coding Experience")
plt.xlabel("Number of Respondents")
plt.ylabel("Years of Experience")
plt.tight_layout()
plt.show()
```

### Job Titles VS. Years of Experience in Coding

```python
job_exp_df = survey_df[['job title','years of experience']].dropna()
top_jobs = survey_df['job title'].value_counts().head().index
filtered = job_exp_df[job_exp_df['job title'].isin(top_jobs)]

plt.figure(figsize=(12,6))
sns.countplot(data=filtered, y='job title', hue='years of experience')
plt.title('Job Titles by Years of Coding Experience (Top 5 Roles)')
plt.xlabel('Number of Respondents')
plt.ylabel('Job Title')
plt.legend(title='Years of Experience')
plt.tight_layout()
plt.show()
```

## `Employment Industry of Kaggle Users`

```python
plt.figure(figsize=(12, 8))
industry_counts = survey_df['industry'].value_counts().head(15)
sns.barplot(x=industry_counts.values, y=industry_counts.index)
plt.title("Employment Industries of Kaggle Users")
plt.xlabel("Number of Respondents")
plt.ylabel("Industry")
plt.tight_layout()
plt.show()
```

## `Salary Distribution of Kaggle Users`

```python
survey_df[['salary']]
```

```python
salary = survey_df[['salary']].iloc[1:].dropna()
salary['salary'].value_counts()
```

```python
salary_count = salary['salary'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=salary_count.values, y=salary_count.index)
plt.title('Yearly Compensation of Kaggle Users (USD)')
plt.xlabel('Number of Respondents')
plt.ylabel('Salary Range')
plt.tight_layout()
plt.show()
```

### Top 5 Salary Ranges

```python
salary_count = salary['salary'].head(10).value_counts().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=salary_count.values, y=salary_count.index)
plt.title('Yearly Compensation of Kaggle Users (USD)')
plt.xlabel('Number of Respondents')
plt.ylabel('Salary Range')
plt.tight_layout()
plt.show()
```

### Salary Ranges by Job Titles of Kaggle Users

```python
salary_job_df = survey_df[['job title', 'salary']].dropna()
salary_job_df
```

```python
salary_job_df = salary_job_df[salary_job_df['job title'] != 'Student']
top_jobs = salary_job_df['job title'].value_counts().head(3).index
filtered = salary_job_df[survey_df['job title'].isin(top_jobs)]

plt.figure(figsize=(12,6))
sns.countplot(data=filtered, y='job title', hue='salary')
plt.title('Salary Distribution by Job Title (Top 3 Roles)')
plt.xlabel('Number of Respondents')
plt.ylabel('Job Title')
plt.legend(title='Salary Range', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.countplot(data=filtered, x='salary', hue='job title')
plt.title('Salary Ranges per Job Title (Top 3 Roles)')
plt.xlabel('Salary Range')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Job Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## `Most Loved Programing Languages`

```python
survey_df.rename(columns = {
    'Q7_Part_1': 'Python',
    'Q7_Part_2': 'R',
    'Q7_Part_3': 'SQL',
    'Q7_Part_4': 'C',
    'Q7_Part_5': 'C++',
    'Q7_Part_6': 'Java',
    'Q7_Part_7': 'Javascript',
    'Q7_Part_8': 'Julia',
    'Q7_Part_9': 'Swift',
    'Q7_Part_10': 'Bash',
    'Q7_Part_11': 'MATLAB',
    'Q7_Part_12': 'None'
}, inplace=True)
```

```python
language_cols = [
    'Python', 'R', 'SQL', 'C', 'C++', 'Java',
    'Javascript', 'Julia', 'Swift', 'Bash',
    'MATLAB', 'None'
]

languages = survey_df.iloc[1:][language_cols].melt(value_name='Language')['Language'].dropna()

language_counts = languages.value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=language_counts.values, y=language_counts.index)
plt.title('Most Used Programming Languages Among Kaggle Users (2021)')
plt.xlabel('Number of Respondents')
plt.ylabel('Programming Language')
plt.tight_layout()
plt.show()
```

### Top 5 Programming Languages

```python
survey_df = survey_df.iloc[1:].reset_index(drop=True)

languages = survey_df[['Python', 'R', 'SQL', 'C', 'C++', 'Java', 'Javascript', 'Julia', 'Swift', 'Bash', 'MATLAB', 'None']].melt(value_name='Language')['Language'].dropna()

language_counts = languages.value_counts().head(5).sort_values(ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(x=language_counts.values, y=language_counts.index)
plt.title('Most Used Programming Languages')
plt.xlabel('Number of Users')
plt.ylabel('Programming Language')
plt.tight_layout()
plt.show()
```
